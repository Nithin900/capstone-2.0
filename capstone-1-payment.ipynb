```python
pv_numeric = pv_care_hos[["Payment", "Lower Estimate", "Higher Estimate"]]
pv_numeric
```

```python
#Payment	Lower Estimate	Higher Estimate
pv_numeric['Payment'] = pv_numeric['Payment'].str.replace('$', '').str.replace(',', '').str.replace("Not Available",'')
pv_numeric['Lower Estimate'] = pv_numeric['Lower Estimate'].str.replace('$', '').str.replace(',', '').str.replace("Not Available",'')
pv_numeric['Higher Estimate'] = pv_numeric['Higher Estimate'].str.replace('$', '').str.replace(',', '').str.replace("Not Available",'')
```

```python
pv_numeric.head(20)
```

```python
pv_numeric.info()
```

```python
pv_numeric["Payment"] = pd.to_numeric(pv_numeric["Payment"], errors='coerce')
pv_numeric["Lower Estimate"] = pd.to_numeric(pv_numeric["Lower Estimate"], errors='coerce')
pv_numeric["Higher Estimate"] = pd.to_numeric(pv_numeric["Higher Estimate"], errors='coerce')
```

```python
pv_numeric.head(20)
```

```python
pv_numeric["Payment"].fillna(pv_numeric["Payment"].mean(), inplace=True)
pv_numeric["Lower Estimate"].fillna(pv_numeric["Lower Estimate"].mean(), inplace=True)
pv_numeric["Higher Estimate"].fillna(pv_numeric["Higher Estimate"].mean(), inplace=True)
```
